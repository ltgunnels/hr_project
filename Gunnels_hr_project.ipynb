## Part 3: Prepare the Data for Modeling

###### Evaluate whether or not there are missing/null values.

Data sets are rarely clean, the code below is necessary to ensure there are no missing values so that we can build correct classification models with this data. It is important to be cleaned up as they could lead to wrong predictions. 

```r
sum(is.na(hr_data))
```

##### Split the DataSet
Split the data into training and test sets. We do this so we can test our prediction model against the actual outcome of the company's attrition.

```r
set.seed(3456)
trainIndex <- createDataPartition(hr_data$Attrition, p = .8, list = FALSE,
                                 times = 1)
hr_train <- hr_data[trainIndex,]
hr_test <- hr_data[-trainIndex,]
```

## Part 4: Build a Classification Model
Using two methods, I am going to build classification models to determine the significant variables affecting attrition. I will use logistic regression first, then decision trees. After building both of those models, I will test the AUC (area under the curve) to prove which model is more accurate and not overfitting to this unique situation. This model might be able to help other datasets and people in the future.  

### Method #1 - Logistic Regression

##### Evaluate the importance of each variable. 
In this section, I am going to run the code multiple times, while eliminating the variables that aren't statistically significant. I will run until all variables are statistically significant (important). Then I will double check with a shortcut to verify I included all the variables. 

```r
# View all the variables and look at p value to determine significance
attrition_model_1 <- glm(Attrition ~ ., data=hr_train)
summary(attrition_model_1)
```

```r
# All these variables remain statistically sifnificant
# I can continue eliminating them to reduce overfitting
attrition_model_2 <- glm(Attrition ~ Age + JobRole + MaritalStatus + OverTime + 
                         BusinessTravel + DistanceFromHome + JobInvolvement+
                       JobSatisfaction + NumCompaniesWorked + YearsInCurrentRole, data=hr_train)
summary(attrition_model_2)
```

```r
# The easy way to verify that the manual model above shows important variables
varImp(attrition_model_2)

importance
```

### Write Predictions
Now that we know which variables we want to use, we can write predicitons with our model from the above codes. We will write these predictions in a new column in the test data. Then I used a function from earlier to view the first few records to ensure that my predictions were inserted into the new column correctly. 

I did this for both the prediction model that focused on a few variables and the model that includes all the variables. I want to test these models against each other to see if I am overfitting. 

* *Overfitting* is when a model is able to fit almost perfectly your training data but is performing poorly on new data. A model will overfit when it is learning the very specific pattern and noise from the training data, this model is not able to extract the “big picture” nor the general pattern from your data.


```r
# Write predictions for my model focused on a few variables
hr_test$attrition_predictions_lr <- predict(attrition_model_2, newdata=hr_test, type="response")
head(hr_test[c("Attrition", "attrition_predictions_lr")])
head(hr_test)
```

```r
# Write predictions for a model including all variables
hr_test$attrition_predictions_lr_all <- predict(attrition_model_1, newdata=hr_test, type="response")
head(hr_test)
```

### Overfitting
Prove we are not overfitting our model by comparing the two models with an AUC model. Since my model is greater than the model including all the variables we see that we are not overfitting. However to improve this data we could eliminate a few more variables from the linear regression model and increase the AUC a little more.

```r
ROC <- roc(hr_test$Attrition, hr_test$attrition_predictions_lr)

plot( ROC, print.auc = TRUE, col = "grey")

ROC2 <- roc(hr_test$Attrition, hr_test$attrition_predictions_lr_all)

plot(ROC2, print.auc = TRUE, col = "black", print.auc.y = .4, add = TRUE)

auc(ROC)
```

### Method #2: Decision Trees
Through this method we will be able to make predictions using decision trees. This is another way to look at the data and see what categories have the biggest impact. We are able to see statistically significant variables and compare this prediction against all the variables and the linear regression example we previously completed. 

We are going to do a similar process to the linear regression model above. We will start by producing a random Forest including all the variables. Then we will show the important variables, and produce a summary of the functions we perfomed to make sure it worked correctly. 

We will then elimate the variables that aren't statistically significant and run the model again with less variables. We will follow the same steps of showing the important variable and producing a summary of the functions we perfomed to make sure it worked correctly. 

```r
# Build a decision tree with all variables
attrition_tree <- randomForest(Attrition ~ ., data=hr_train, importance=TRUE)

varImpPlot(attrition_tree)
summary(attrition_tree)
```

```r
# Build a decision tree with only a few variables selected
attrition_tree_2 <- randomForest(Attrition ~ Age + JobRole + MaritalStatus + OverTime + 
                         BusinessTravel + DistanceFromHome + JobInvolvement+
                       JobSatisfaction + NumCompaniesWorked + YearsInCurrentRole, data=hr_train, importance=TRUE)

varImpPlot(attrition_tree_2)
summary(attrition_tree_2)
```

### Write Predictions

Next we will again enter the predictions from this new model into a new column in the test data called "attrition_tree_pred_all" and "attrition_tree_pred_2." We will use the head function to show the first few records and view the new column we created. 

```r
# Write new column into test data with my predicitons from the first model with all variables
hr_test$attrition_tree_pred_all <- predict(attrition_tree, newdata=hr_test, type="response")
head(hr_test)
```

```r
# Write new column with my predictions from the second model with selected variables
hr_test$attrition_tree_pred_2 <- predict(attrition_tree_2, newdata=hr_test, type="response")
head(hr_test)
```

### Overfitting
We will again compare the models and evaluate the performance of each. This time we will use one linear regression model and one random Forest model. From the AUC graph we see that the linear regression model is a better method and therefore we will use those predictions in our final report. 

```r
ROC <- roc(hr_test$Attrition, hr_test$attrition_tree_pred_2)

plot( ROC, print.auc = TRUE, col = "black")

ROC2 <- roc(hr_test$Attrition, hr_test$attrition_predictions_lr)

plot(ROC2, print.auc = TRUE, col = "grey", print.auc.y = .4, add = TRUE)

auc(ROC)
```
